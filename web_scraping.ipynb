```python
 import requests
 from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.amazon.in/Samsung-Midnight-Storage-sAMOLED-Display/product-reviews/B0C7BZX934/ref=cm_cr_othr_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(url)
page
# output200 means all good
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
soup
```

```python
name = soup.find_all('span', class_ = 'a-profile-name')
name
# with[2;] we removed first 2 reveiws which repeated
```

```python
name = name[2:]
name
```

```python
len(name)
```

```python
import pandas as pd
```

```python
 a = ['customer name']
```

```python
df= pd.DataFrame(name, columns = a)
```

```python
df
```

```python
DF = df.drop_duplicates(ignore_index= True)
```

```python
DF
```

```python
r_title= soup.find_all('a', class_ = 'review-title')
r_title
```

```python
len(r_title)
```

```python
review_title = []
for i in range(0, len(r_title)):
   review_title.append(r_title[i].get_text()[19:-1])
review_title
```

```python
len(review_title)
```

```python
# review dates
r_dates= soup.find_all('span',class_ = 'review-date')[2:]
r_dates
```

```python
len(r_dates)
```

```python
date = []
for i in range(0, len(r_dates)):
   date.append(r_dates[i].get_text()[20:])
date
```

```python
#review rating in stars
r_stars= soup.find_all('i',class_ = 'review-rating')[2:]
r_stars
```

```python
rating = []
for i in range(0, len(r_stars)):
   rating.append(r_stars[i].get_text()[0:])
rating
```

```python
r_content= soup.find_all('span', class_ = 'review-text')
r_content
```

```python
len(r_content)
```

```python
content = []
for i in range(0, len(r_content)):
   content.append(r_content[i].get_text()[1:-1])
content
```

```python
DF['Review Title']= review_title
DF['Review Date']= date
DF['Review Rating']= rating
DF['Review Content']= content
```
